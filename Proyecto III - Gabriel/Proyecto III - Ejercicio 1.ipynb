 Vaciar los datos del fichero barca.csv en un dataframe. Debe mostrar los encabezados de los datos y
describir la cantidad de alojamientos disponibles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
#Comprueba la carpeta de trabajo actual

PATH_FILE = os.getcwd()
PATH_DATA = PATH_FILE + "\\data\\"

# Cargar el archivo CSV en un DataFrame
data = pd.read_csv(PATH_DATA+'barca.csv')

# Mostrar los encabezados de los datos
print(data.head())

# Contar y mostrar la cantidad de alojamientos disponibles
num_alojamientos = len(data)
print("Cantidad de alojamientos disponibles:", num_alojamientos)
```

Mostrar la información del dataset, debe realizar una breve descripción de los datos nombre, tipo de
datos y si existen valores nulos, cantidad de atributos numéricos y categóricos (objetos o texto)

```python


# Mostrar la información del dataset
print("Información del dataset:")
print(data.info())

# Contar y mostrar la cantidad de atributos numéricos
num_numeric = len(data.select_dtypes(include=['number']).columns)
print("Cantidad de atributos numéricos:", num_numeric)

# Contar y mostrar la cantidad de atributos categóricos (objetos o texto)
num_categorical = len(data.select_dtypes(include=['category']).columns) + len(data.select_dtypes(include=['object']).columns)
print("Cantidad de atributos categóricos:", num_categorical)
```

Mostrar los datos estadísticos del dataset, describa los resultados obtenidos para Price

```python


# Mostrar los datos estadísticos del dataset
print("Datos estadísticos del dataset:")
print(data.describe())

# Describir los resultados obtenidos para la columna 'Price'
price_stats = data['price'].describe()
print("\nResultados obtenidos para la columna 'Price':")
print(price_stats)
```

Mostrar la cantidad de alojamientos que existe para cada tipo de alojamiento (room_type) cantidad
y porcentaje

```python


# Contar la cantidad de alojamientos por tipo
room_type_counts = data['room_type'].value_counts()
print("Cantidad de alojamientos por tipo:")
print(room_type_counts)

# Calcular y mostrar los porcentajes de alojamientos por tipo
room_type_percentages = data['room_type'].value_counts(normalize=True) * 100
print("\nPorcentaje de alojamientos por tipo:")
print(room_type_percentages)
```

Mostrar un gráfico % alojamientos que existe para cada tipo de alojamiento (room_type)

```python
# Calcular los porcentajes de alojamientos por tipo
room_type_percentages = data['room_type'].value_counts(normalize=True) * 100

# Crear un gráfico de barras
plt.figure(figsize=(8, 6))
room_type_percentages.plot(kind='bar')
plt.xlabel('Tipo de alojamiento')
plt.ylabel('Porcentaje de alojamientos')
plt.title('Porcentaje de alojamientos por tipo de alojamiento')
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

Mostrar un gráfico de dispersión del atributo precio

```python
# Crear un gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(data.index, data['price'], alpha=0.5)
plt.xlabel('Índice')
plt.ylabel('Precio')
plt.title('Gráfico de dispersión del atributo precio')

# Mostrar el gráfico
plt.show()
```

Mostrar un gráfico de caja (boxplot) para precio, habilite la opción para visualizar los datos atípicos.

```python
# Crear un gráfico de caja
plt.figure(figsize=(8, 6))
plt.boxplot(data['price'], vert=True, showfliers=True)
plt.xlabel('Precio')
plt.title('Gráfico de caja (boxplot) para el atributo precio con datos atípicos')

# Mostrar el gráfico
plt.show()

print("Visualmente se puede intuir que las datos recibidos no siguen una distribución normal \n Ya que los datos no son simétricos respecto al eje de la caja y ademas hay muchos outliers")
```

Mostrar la cantidad de alojamientos por barrio "neighborhood"

```python
# Contar la cantidad de alojamientos por barrio
neighborhood_counts = data['neighborhood'].value_counts()
print("Cantidad de alojamientos por barrio:")
print(neighborhood_counts)
```

Mostrar un gráfico que muestre los tipos de alojamiento (room_type) por barrio "neighborhood"

```python
# Crear un gráfico de barras con seaborn
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='neighborhood', hue='room_type')
plt.xlabel('Barrio')
plt.ylabel('Cantidad de alojamientos')
plt.title('Tipos de alojamiento por barrio')
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

Mostrar los histogramas de los datos numéricos del dataset

```python
# Seleccionar solo las columnas numéricas del dataset
numeric_data = data.select_dtypes(include=['number'])

# Crear histogramas de los datos numéricos
numeric_data.hist(figsize=(12, 8), bins=20)
plt.tight_layout()

# Mostrar los histogramas
plt.show()
```

 Mostrar la matriz de correlación de los datos numéricos

```python
# Calcular la matriz de correlación
correlation_matrix = numeric_data.corr()

# Crear un mapa de calor (heatmap) para visualizar la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlación de los datos numéricos')


# Rotar los nombres de los ejes en un ángulo de 45 grados
plt.xticks(rotation=45, ha='right')


# Mostrar el gráfico
plt.show()

print("Los 3 pares más relacionados son: \n bedrooms - accommodates \n price - accommodates \n price - bedrooms")
```

Mostrar un gráfico 3D que muestre la ubicación (latitud, longitud) de los tres tipos de alojamientos

```python
# Crear una figura y un eje 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Crear un gráfico de dispersión 3D para cada tipo de alojamiento
room_types = data['room_type'].unique()
colors = ['r', 'g', 'b']

for room_type, color in zip(room_types, colors):
    selected_data = data[data['room_type'] == room_type]
    ax.scatter(selected_data['latitude'], selected_data['longitude'], zs=0, zdir='z', s=20, c=color, label=room_type)

# Etiquetas y título
ax.set_xlabel('Latitud')
ax.set_ylabel('Longitud')
ax.set_zticks([])
ax.set_title('Ubicación (latitud, longitud) de los tres tipos de alojamientos')
ax.legend()

# Mostrar el gráfico
plt.show()
```

Mostrar un grafico que muestre la ubicación de los alojamientos (latitud, longitud), utilice una
columna de color (colobar) para representar los precios de los alojamientos

```python
# Crear una figura
plt.figure(figsize=(10, 8))

# Crear un gráfico de dispersión con una columna de color (colorbar) para representar los precios
sc = plt.scatter(data['longitude'], data['latitude'], c=data['price'], cmap='viridis', s=20, alpha=0.7)
plt.colorbar(sc, label='Precio')

# Etiquetas y título
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Ubicación de los alojamientos y precios')

# Mostrar el gráfico
plt.show()
```
